````python
from ultralytics import YOLO

# YOLOv8 modelini yükle
model = YOLO("yolov8n.pt")  # YOLOv8 nano (hafif ve hızlı model)

# Modeli eğit
results = model.train(
    data="/content/dataset/data.yaml",  # YAML dosyanın yolunu kontrol et
    epochs=50,   # Epoch sayısını belirle (50 iyi bir başlangıçtır)
    batch=16,    # Batch size (GPU'na göre artırılabilir)
    imgsz=640,   # Görsel boyutu (YOLO genellikle 640 kullanır)
    device="cuda"  # GPU kullanımı (Eğer CPU kullanıyorsan "cpu" yaz)
)
````

````python
import os
import random
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Modeli yükle
model = YOLO("/content/runs/detect/train/weights/best.pt")

# Test klasörünü tanımla
test_folder = "/content/dataset/test/images"

# Test klasöründeki tüm görüntüleri listele
image_files = [f for f in os.listdir(test_folder) if f.endswith((".jpg", ".png", ".jpeg"))]

# Rastgele bir görüntü seç
random_image = random.choice(image_files)
image_path = os.path.join(test_folder, random_image)

print(f"Test edilen görüntü: {random_image}")

# Modeli çalıştır
results = model.predict(image_path, save=True)

# Görüntüyü yükle ve sonucu göster
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV BGR formatında okuyor, RGB'ye çevirelim

# Tahmin edilen görüntüyü çizelim
plt.figure(figsize=(10, 6))
plt.imshow(img)
plt.axis("off")
plt.title("Model Tahmin Sonucu")
plt.show()
````
